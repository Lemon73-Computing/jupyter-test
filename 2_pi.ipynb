## Finding pi

```csharp
Random rand = new();
int x = 0, y = 0;

// How many times do you try?
int times = 5000;
// How long harf range of circle?
int r = 100;

int clear = 0;
for (int i = 0;i < times;i++){
    x = rand.Next(0, r + 1);
    y = rand.Next(0, r + 1);
    double root = Math.Sqrt(Math.Pow(x, 2) + Math.Pow(y, 2));
    if (r >= root)
        clear++;
    // Console.WriteLine(x.ToString() + " " + y.ToString() + " " + root.ToString() + " " + clear.ToString());
}

double pi = (double)(times - clear) / times;

Console.WriteLine(pi);
```
